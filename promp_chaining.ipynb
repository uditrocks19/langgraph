```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict
import os
```

```python
model = ChatOpenAI(api_key=os.getenv('API_KEY'))
```

```python
class LLMState(TypedDict):
    topic: str
    outline: str
    content: str
```

```python
graph = StateGraph(LLMState)
```

```python
def create_outline(state: LLMState) -> LLMState:
    topic = state['topic']
    prompt = f"Create a detailed outline for an article about {topic}."
    outline = model.invoke(prompt).content
    state['outline'] = outline
    return state
```

```python
def create_content(state: LLMState) -> LLMState:
    outline = state['outline']
    prompt = f"Write a detailed article based on the following outline:\n{outline}"
    content = model.invoke(prompt).content
    state['content'] = content
    return state
```

```python
graph.add_node('generate_outline', create_outline)
graph.add_node('generate_content', create_content)

graph.add_edge(START, 'generate_outline')
graph.add_edge('generate_outline', 'generate_content')
graph.add_edge('generate_content', END)

wf = graph.compile()
```

```python
# initial state

initial_state = {'topic': 'The impact of AI on society'}
final_state = wf.invoke(initial_state)
final_state
```
